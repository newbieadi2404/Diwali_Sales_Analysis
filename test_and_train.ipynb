```python
from sklearn.preprocessing import OrdinalEncoder

data_encoded = data.copy()

cat_cols = data.select_dtypes(include=['object']).columns
for col in cat_cols:
    oe = OrdinalEncoder()
    data_encoded[col]=oe.fit_transform(data_encoded[[col]])
    print(oe.categories_)
```

```python
data_encoded.head(20)
```

```python
X = data_encoded.drop('Occupation', axis=1) 
Y = data_encoded['Occupation']
```

```python
Y
```

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import cross_val_score

rf = RandomForestRegressor(n_estimators=100,random_state=42)
scores = cross_val_score(rf,X,Y,cv=5,scoring='r2')
print(scores.mean())
```

```python
import tkinter as tk
from tkinter import ttk
from IPython.display import display

# Create a Tkinter window
window = tk.Tk()
window.title("Diwali Sales Data GUI")

# Create labels and entries for each field
user_id_label = tk.Label(window, text="User ID:")
user_id_label.grid(row=0, column=0, padx=5, pady=5)
user_id_entry = tk.Entry(window)
user_id_entry.grid(row=0, column=1, padx=5, pady=5)

cust_name_label = tk.Label(window, text="Customer Name:")
cust_name_label.grid(row=1, column=0, padx=5, pady=5)
cust_name_entry = tk.Entry(window)
cust_name_entry.grid(row=1, column=1, padx=5, pady=5)

product_id_label = tk.Label(window, text="Product ID:")
product_id_label.grid(row=2, column=0, padx=5, pady=5)
product_id_entry = tk.Entry(window)
product_id_entry.grid(row=2, column=1, padx=5, pady=5)

gender_label = tk.Label(window, text="Gender:")
gender_label.grid(row=3, column=0, padx=5, pady=5)
gender_var = tk.StringVar()
gender_var.set("Male")  # default value
gender_option = tk.OptionMenu(window, gender_var, "Male", "Female")
gender_option.grid(row=3, column=1, padx=5, pady=5)

age_group_label = tk.Label(window, text="Age Group:")
age_group_label.grid(row=4, column=0, padx=5, pady=5)
age_group_var = tk.StringVar()
age_group_var.set("18-24")  # default value
age_group_option = tk.OptionMenu(window, age_group_var, "18-24", "25-34", "35-44", "45-54", "55-64", "65+")
age_group_option.grid(row=4, column=1, padx=5, pady=5)

age_label = tk.Label(window, text="Age:")
age_label.grid(row=5, column=0, padx=5, pady=5)
age_entry = tk.Entry(window)
age_entry.grid(row=5, column=1, padx=5, pady=5)

marital_status_label = tk.Label(window, text="Marital Status:")
marital_status_label.grid(row=6, column=0, padx=5, pady=5)
marital_status_var = tk.StringVar()
marital_status_var.set("Single")  # default value
marital_status_option = tk.OptionMenu(window, marital_status_var, "Single", "Married", "Divorced", "Widowed")
marital_status_option.grid(row=6, column=1, padx=5, pady=5)

state_label = tk.Label(window, text="State:")
state_label.grid(row=7, column=0, padx=5, pady=5)
state_entry = tk.Entry(window)
state_entry.grid(row=7, column=1, padx=5, pady=5)

zone_label = tk.Label(window, text="Zone:")
zone_label.grid(row=8, column=0, padx=5, pady=5)
zone_entry = tk.Entry(window)
zone_entry.grid(row=8, column=1, padx=5, pady=5)

occupation_label = tk.Label(window, text="Occupation:")
occupation_label.grid(row=9, column=0, padx=5, pady=5)
occupation_entry = tk.Entry(window)
occupation_entry.grid(row=9, column=1, padx=5, pady=5)

product_category_label = tk.Label(window, text="Product Category:")
product_category_label.grid(row=10, column=0, padx=5, pady=5)
product_category_entry = tk.Entry(window)
product_category_entry.grid(row=10, column=1, padx=5, pady=5)

orders_label = tk.Label(window, text="Orders:")
orders_label.grid(row=11, column=0, padx=5, pady=5)
orders_entry = tk.Entry(window)
orders_entry.grid(row=11, column=1, padx=5, pady=5)

amount_label = tk.Label(window, text="Amount:")
amount_label.grid(row=12, column=0, padx=5, pady=5)
amount_entry = tk.Entry(window)
amount_entry.grid(row=12, column=1, padx=5, pady=5)

# Create a submit button
def submit_data():
    user_id = user_id_entry.get()
    cust_name = cust_name_entry.get()
    product_id = product_id_entry.get()
    gender = gender_var.get()
    age_group = age_group_var.get()
    age = age_entry.get()
    marital_status = marital_status_var.get()
    state = state_entry.get()
    zone = zone_entry.get()
    occupation = occupation_entry.get()
    product_category = product_category_entry.get()
    orders = orders_entry.get()
    amount = amount_entry.get()
    
    # Print the data to the console
    print("User ID:", user_id)
    print("Customer Name:", cust_name)
    print("Product ID:", product_id)
    print("Gender:", gender)
    print("Age Group:", age_group)
    print("Age:", age)
    print("Marital Status:", marital_status)
    print("State:", state)
    print("Zone:", zone)
    print("Occupation:", occupation)
    print("Product Category:", product_category)
    print("Orders:", orders)
    print("Amount:", amount)

submit_button = tk.Button(window, text="Submit", command=submit_data)
submit_button.grid(row=13, column=0, columnspan=2, padx=5, pady=5)

# Start the Tkinter event loop
window.mainloop()
```
